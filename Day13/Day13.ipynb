```python
import numpy as np
```

```python
def piego_y( array, y ):
    
    nrow = int(len(array[:,0]))
    ncol = int(len(array[0,:]))
    for i in range(y+1, nrow):
        for j in range(ncol):
            if array[i,j] == 1:
                array[i,j] = 0
                idx = abs(i-y)
                array[y-idx, j] = 1
    return array[:y,:]
def piego_x( array, x ):
    
    nrow = int(len(array[:,0]))
    ncol = int(len(array[0,:]))
    for i in range( nrow):
        for j in range(x+1,ncol):
            if array[i,j] == 1:
                array[i,j] = 0
                idx = abs(j-x)
                array[i, x-idx] = 1
    return array[:,:x]   

f = open('day13.txt', 'r')
Lines = f.read().splitlines()
l = []
fold_x = []
fold_y = []
for line in Lines:
    if "," in line:
        l.append( line.split(",") )
    if "x" in line:
        fold_x.append(line.split("=")[1]   )
    if "y" in line:
        fold_y.append(line.split("=")[1]   )
print( fold_x)
print( fold_y)
```

```python
coor = np.array(l, dtype = 'int')

nrow = np.max(coor[:,0])
ncol = np.max(coor[:,1])
print( nrow, ncol)
points = np.zeros((ncol+1, nrow+1))
for k in range(int(len(coor))):
    i = coor[k,0]
    j = coor[k,1]
    points[j,i] += 1
print( points)
print( points[3,0])
```

```python
f = open('day13.txt', 'r')
Lines = f.read().splitlines()
k = 0
for line in Lines:
    if "=" in line:
        if "x" in line:
            k = k + 1
            x = int( line.split("=")[1] )
            points = piego_x( points, int(x)  )
        if "y" in line:
            k = k +1 
            y = int( line.split("=")[1] )
            points = piego_y( points, int(y)  )

print( points.sum() )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for k in range(8):
    #print( points[:,k*5:(k+1)*5])
    sns.heatmap(points[:,k*5:(k+1)*5]) 
    plt.show()
    plt.close()
```
